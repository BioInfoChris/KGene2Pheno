```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np 

# Import the CSV file as a DataFrame
datapath = '/home/antoine/gene_pheno_pred/figures/wandb_trialrun.csv'
df = pd.read_csv(datapath)
df['MRR'] = df[[f'Hit@{k}' for k in range(1, 11)]].mean(axis=1)
```

# TransE

```python
# Filter to only conver ConvKB methods
data = df[df['method'] == 'TransE']
x = data["n_epochs"]
y = data["batch_size"]

for metric in ["Hit@1", "MRR"]:
    # plot the trisurf
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_trisurf(data['n_epochs'], data['batch_size'], data[metric], cmap='Blues', alpha=0.9)
    ax.set_xlabel('n_epochs')
    ax.set_ylabel('batch_size')
    ax.set_zlabel('Hit@1')
    ax.set_title(f'TransE - Embedding quality ({metric})')
    ax.set_zlim3d(0, 1)
    ax.set_zlim(0, 1)

plt.show()
```

# ConvKB

```python
# Filter to only conver ConvKB methods
tmp = df[df['method'] == 'ConvKB']

# Create separate DataFrames for each combination of loss function and number of filters
losses = tmp['loss_fn'].unique()
filters = tmp['n_filters'].unique()
dfs = {}

for loss in losses:
    for filt in filters:
        key = f'{loss}_{filt}'
        dfs[key] = tmp[(tmp['loss_fn'] == loss) & (tmp['n_filters'] == filt)]
```

```python
# Set up the figure and axes
fig = plt.figure(figsize=(15, 15))
axes = fig.subplots(nrows=len(losses), ncols=len(filters), subplot_kw={'projection': '3d'})

# Loop over the combinations of loss function and number of filters
for i, loss in enumerate(losses):
    for j, filt in enumerate(filters):
        key = f'{loss}_{filt}'
        data = dfs[key]
        
        # Create the 3D plot
        ax = axes[i, j]

        ax.scatter(data['n_epochs'], data['batch_size'], data['Hit@1'])
        ax.set_xlabel('n_epochs')
        ax.set_ylabel('batch_size')
        ax.set_zlabel('Hit@1')
        ax.set_title(key)

        # Set the axes at scale with each other
        ax.set_zlim3d(0, 1)
# Show the plots
plt.show()
```

```python
df = pd.read_csv("wandb.csv")
# plot showing how batch size affects runtime
df.groupby('batch_size')['Runtime'].mean().plot(kind='bar')
plt.xlabel('Batch Size')
plt.ylabel('Runtime (sec)')
plt.title('Effect of Batch Size on Runtime')
plt.show()
```

```python
# plot showing how epoch affects runtime
df.groupby('n_epochs')['Runtime'].mean().plot(kind='bar')
plt.xlabel('Number of Epochs')
plt.ylabel('Runtime (sec)')
plt.title('Effect of Epochs on Runtime')
plt.show()
```

```python
# plot showing the effect of method on runtime
df.groupby('method')['Runtime'].mean().plot(kind='bar')
plt.xlabel('Method')
plt.ylabel('Runtime (sec)')
plt.title('Mean runtime per method')
plt.show()

```
