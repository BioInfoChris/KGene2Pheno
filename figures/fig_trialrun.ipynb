```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np 

# Import the CSV file as a DataFrame
datapath = '/home/antoine/gene_pheno_pred/figures/wandb_trialrun.csv'
df = pd.read_csv(datapath)
df['MRR'] = df[[f'Hit@{k}' for k in range(1, 11)]].mean(axis=1)
```

# TransE

```python
data = df[df['method'] == 'TransE']
data = data.drop(data[data['dissimilarity_type'].str.contains('L2')].index)
data = data.drop(data[data['ent_emb_dim'] != 50].index)
data = data.drop(data[data['margin'] != 1].index)

# plot the Hit@1 trisurf
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d', computed_zorder=False)
ax.plot_trisurf(data['n_epochs'], data['batch_size'], data['Hit@1'], cmap='Blues', alpha=0.9, zorder=1)
ax.set_xlabel('n_epochs')
ax.set_ylabel('batch_size')
ax.set_zlabel('Hit@1')
ax.set_title(f'TransE - Embedding quality (Hit@1)')
ax.set_zlim3d(0, 1)
ax.set_zlim(0, 1)
max_idx = data['Hit@1'].idxmax()
max_x = data['n_epochs'][max_idx]
max_y = data['batch_size'][max_idx]
max_z = data['Hit@1'][max_idx]
ax.scatter(max_x, max_y, max_z, c='r', marker='o', s=100)
ax.text(max_x + 0.5, max_y + 300, max_z, f'{max_z:.3f}', color='red')
plt.show()

# plot the MRR trisurf
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d', computed_zorder=False)
ax.plot_trisurf(data['n_epochs'], data['batch_size'], data['MRR'], cmap='Blues', alpha=0.9, zorder=1)
ax.set_xlabel('n_epochs')
ax.set_ylabel('batch_size')
ax.set_zlabel('MRR')
ax.set_title(f'TransE - Embedding quality (MRR)')
ax.set_zlim3d(0, 1)
ax.set_zlim(0, 1)
max_idx = data['Hit@1'].idxmax()
max_x = data['n_epochs'][max_idx]
max_y = data['batch_size'][max_idx]
max_z = data['MRR'][max_idx]
ax.scatter(max_x, max_y, max_z, c='r', marker='o', s=100)
ax.text(max_x + 0.5, max_y + 300, max_z, f'{max_z:.3f}', color='red')
plt.show()
```

# ConvKB

```python
# Filter to only conver ConvKB methods
tmp = df[df['method'] == 'ConvKB']
tmp = tmp.drop(tmp[tmp['dissimilarity_type'].str.contains('L2')].index)
tmp = tmp.drop(tmp[tmp['ent_emb_dim'] != 50].index)
tmp = tmp.drop(tmp[tmp['margin'] != 1].index)

# Create separate DataFrames for each combination of loss function and number of filters
losses = tmp['loss_fn'].unique()
filters = tmp['n_filters'].unique()
dfs = {}

for loss in losses:
    for filt in filters:
        key = f'{loss}_{filt}'
        dfs[key] = tmp[(tmp['loss_fn'] == loss) & (tmp['n_filters'] == filt)]
```

```python


# Loop over the combinations of loss function and number of filters
for i, loss in enumerate(losses):
    for j, filt in enumerate(filters):
        key = f'{loss}_{filt}'
        data = dfs[key]


        if len(data['n_epochs']) >= 3 and len(data['batch_size']) >= 3 and len(data['Hit@1']) >= 3:
            # Create a new figure and axes
            fig = plt.figure()
            ax = fig.add_subplot(111, projection='3d', computed_zorder=False)

            # Create the 3D plot
            ax.plot_trisurf(data['n_epochs'], data['batch_size'], data['Hit@1'], cmap='Blues', alpha=0.9, zorder=1)
            ax.set_xlabel('n_epochs')
            ax.set_ylabel('batch_size')
            ax.set_zlabel('Hit@1')
            ax.set_title(f'ConvKB - Embedding quality (Hit@1, {loss}, {filt})')

            
            # Set the axes at scale with each other
            # ax.set_xlim3d((0, 50))
            # ax.set_ylim3d((0, 11000))
            ax.set_zlim3d(0, 1)

            # Add a mark for the highest point in the Z axis, in front of the plot
            max_idx = data['Hit@1'].idxmax()
            max_x = data['n_epochs'][max_idx]
            max_y = data['batch_size'][max_idx]
            max_z = data['Hit@1'][max_idx]
            ax.scatter(max_x, max_y, max_z, c='r', marker='o', s=100, zorder=10)
            ax.text(max_x + 0.9, max_y, max_z + 0.02, f'{max_z:.3f}', color='red')

            # Show the plot for this combination of loss function and number of filters
            plt.show()
```
