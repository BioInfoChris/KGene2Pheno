```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
model_order = [
    'Random Forest', 'Extra Trees Classifier', 'Light Gradient Boosting Machine', 'MLP Classifier','Decision Tree Classifier','K Neighbors Classifier', 'Linear Discriminant Analysis','Ridge Classifier', 'SVM - Linear Kernel','Quadratic Discriminant Analysis', 'Naive Bayes']
     
```

```python
# TransE

models = ['K Neighbors Classifier', 'Naive Bayes', 'Decision Tree Classifier', 'SVM - Linear Kernel',
          'Ridge Classifier', 'Random Forest', 'Quadratic Discriminant Analysis', 'Linear Discriminant Analysis',
          'Extra Trees Classifier', 'Light Gradient Boosting Machine', 'MLP Classifier']
accuracy = [0.7096, 0.6231, 0.8155, 0.6972, 0.7076, 0.8644, 0.6287, 0.7076, 0.8614, 0.8534, 0.8163]
accuracy_std = [0.0029, 0.0029, 0.0025, 0.0042, 0.0031 ,0.0014, 0.0028, 0.0031, 0.0015, 0.0015, 0.0021]
recall = [0.6299, 0.5273, 0.8221, 0.8407, 0.8068, 0.9130, 0.5552, 0.8068, 0.8942, 0.9087, 0.8389]
recall_std = [0.0051, 0.0053, 0.0038, 0.0038, 0.0025 ,0.0019, 0.0062, 0.0025, 0.0019, 0.0033, 0.0066]
precision = [0.7494, 0.6523, 0.8114, 0.6533, 0.6732 ,0.8321, 0.6509, 0.6732 ,0.8392, 0.8182, 0.8027]
precision_std = [0.0034, 0.0041, 0.0018, 0.0038, 0.0033 ,0.0021, 0.0033, 0.0033, 0.0019, 0.0015, 0.0029]
transe = pd.DataFrame({'Model': models, 'Accuracy': accuracy, 'sd': accuracy_std, 'Recall': recall, 'rec_sd': recall_std, 'Precision': precision, 'prec_sd': precision_std})
transe = transe.sort_values(by='Model', key=lambda x: x.map({m: i for i, m in enumerate(model_order)}))
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
sns.barplot(x='Accuracy', y='Model', data=transe, ax=ax, errorbar='sd')
ax.set_title('Accuracy of binary classification models', size=18)
ax.set_xlabel('Accuracy', size=18)
ax.set_ylabel('TransE', size=18)
ax.set_xlim(0.5, 1)
# Increase font size of tick labels
ax.tick_params(axis='both', labelsize=14)
ax.errorbar(x=transe['Accuracy'], y=transe['Model'] ,xerr=transe['sd'], fmt='none', ecolor='black')
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(9, 6))
sns.barplot(x='Precision', y='Model', data=transe, ax=ax, errorbar='sd')
ax.set_title('', size=18)
ax.set_xlabel('Precision', size=18)
ax.set_ylabel('TransE', size=18)
ax.set_xlim(0.5, 1)
# Increase font size of tick labels
ax.tick_params(axis='both', labelsize=14)
ax.errorbar(x=transe['Precision'], y=transe['Model'], xerr=transe['prec_sd'], fmt='none', ecolor='black')
plt.tight_layout()
plt.show()
```
